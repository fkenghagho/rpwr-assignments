### Task 2 - URDF ( 3 Pts )

In Assignment2, you made use of visualization markers to show your turtle robot in rviz2. However, this solution is brittle and a much cleaner solution would be to create the URDF description of your turtle robot, publish to rviz2 and then set the base frame of your robot to turtle1 so that by moving turtle1 your visualization will also move. **Provide your solution in my_robot.urdf.**

**NB: To validate your construction, run the followwing commands:**
- Make sure the robot and joint state publisher gui package are installed:
  ```
  sudo apt update
  sudo apt install ros-jazzy-joint-state-publisher-gui
  sudo apt install ros-jazzy-robot-state-publisher
  ```
- Publish the urdf: ros2 run robot_state_publisher robot_state_publisher my_robot.urdf
- Create a joint state publisher: ros2 run joint_state_publisher_gui joint_state_publisher_gui
- Run the turtle demo to see if your model moves with the turtle.
- Make sure that there is a transfor between your base link to the world/map link
- Start rviz2 and play with the joint state publisher
- The code above highlight a simple two-link one-joint robot arm.
```
<?xml version="1.0"?>
<robot name="simple_two_link_arm">

  <link name="base_link">
    <visual>
      <geometry>
        <cylinder radius="0.2" length="0.1"/>
      </geometry>
      <material name="blue">
        <color rgba="0 0 0.8 1"/>
      </material>
    </visual>
  </link>

  <joint name="base_to_arm_joint" type="continuous">
    <parent link="base_link"/>
    <child link="arm_link"/>
    <origin xyz="0 0 0.05" rpy="0 0 0"/>
    <axis xyz="0 0 1"/> </joint>

  <link name="arm_link">
    <visual>
      <origin xyz="0 0 0.25" rpy="0 0 0"/>
      <geometry>
        <box size="0.1 0.1 0.5"/>
      </geometry>
      <material name="green">
        <color rgba="0 0.8 0 1"/>
      </material>
    </visual>
  </link>

</robot>
```
